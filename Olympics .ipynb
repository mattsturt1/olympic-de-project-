```python
#%matplotlib notebook

# Dependencies
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Olympic_data = pd.read_csv("athlete_events.csv")
Olympic_data.head()
```

```python
Olympic_data_df = Olympic_data.loc[:, ["Sex", "Team", "Year",
                                          "Games", "Season", "Sport", "NOC", "Weight"]]
Olympic_data_df
```

```python
Olympic_data_df["Sport"].value_counts()
```

```python
Sports = ["Athletics", "Gymnastics", "Swiming", "Shooting", "Cycling"]
Total_Entreis = [ 38624,26707, 23195, 11448, 10859]
x_axis = np.arange(len(Total_Entreis))
```

```python
plt.bar(x_axis, Total_Entreis, color="b", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, Sports)
plt.title("Top 5 Sports by Participants")
plt.xlabel("Sports")
plt.ylabel("Participants")
plt.show()
```
